# Subgraficación (Subploting)

A continuación se muestra una forma de realizar la subgraficación.

O la graficación de varias funciones en un solo gráfico generado
por medio de la función plot

```julia
x
```

```julia
p1 = plot(x,x)
p2 = plot(x,x.^2)
p3 = plot(x,x.^3)
p4 = plot(x,x.^4)
plot(p1,p2,p3,p4,layout=(2,2),legend=false)
```

```julia
p1 = plot(x,x)
p2 = plot(x,x.^2)
p3 = plot(x,x.^3)
p4 = plot(x,x.^4)
plot(p1,p2,p3,p4,layout=(4,1),legend=false)
```
